Importing all required modules

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
plt.style.use('dark_background')
```

Loading the Dataset 

```python
df=pd.read_csv("zomato.csv\zomato.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

Deleting the unwanted Values(Columns)

```python
df=df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis=1)
df
```

Checking the presence of null Values

```python
df.info()
```

Deleting the Duplicate Values from the dataset

```python
df.drop_duplicates(inplace=True)
df.shape
```

Cleaning Every Column

```python
#checking the rate column to see the data values present in the dataset
df['rate'].unique()
```

To remove the string values we're gonna replace with NaN

```python
def removeSval(keyVal):
    if(keyVal=="NEW" or keyVal=="-"):
        return np.nan
    else:
        keyVal=str(keyVal).split("/")[0]
        return float(keyVal)
df['rate']=df['rate'].apply(removeSval)
df['rate'].head()
```

Now Calclating the Null Values present in the dataset

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
df.rate.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2persons','listed_in(type)':'type'},inplace=True)
```

```python
df
```

```python
df
```

As both the location column and listed in city(contains the same data entries removing the column listed-in(city))

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df
```

```python
df['cost2persons'].unique()
```

The Values of Cost needs to be either in integer or float but here in the dataset it is present in string hence the commas are removed and returned as a float value

```python
def removeComma(value):
    value=str(value)
    if "," in value:value=value.replace(",","")
    return float(value)
df['cost2persons']=df['cost2persons'].apply(removeComma)
df['cost2persons'].unique()
```

```python
df.head()
```

Now onto the rest_type column

```python
restType=df['rest_type'].value_counts(ascending=False)
restType
```

Clustering the Dessert , Parlor ,...... and other category of rest types as other 

```python
restTypeLessThan1000=restType[restType<1000]
restTypeLessThan1000
```

Clustering part goes here 

```python
def restTypeFunc(value):
    if value in restTypeLessThan1000:return "others"
    else:return value
df['rest_type']=df['rest_type'].apply(restTypeFunc)
df['rest_type'].value_counts()
```

Now on to clean the location column

```python
df['location'].value_counts()
```

As Previously done in the rest type column the location is also clustered as other location which has value less than 300

```python
location=df['location'].value_counts(ascending=False)
locationLessThan300=location[location<300]
locationLessThan300
```

```python
def locationFunc(value):
    if value in locationLessThan300:return "others"
    else:return value
df['location']=df['location'].apply(locationFunc)
df['location'].value_counts()
```

Now on to the cuisine column

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisinesLessThan100=cuisines[cuisines<100]
cuisinesLessThan100
```

```python
def cuisinesFunc(value):
    if value in cuisinesLessThan100:
        return "others"
    else:
        return value
df['cuisines']=df['cuisines'].apply(cuisinesFunc)
df['cuisines'].value_counts()
```
